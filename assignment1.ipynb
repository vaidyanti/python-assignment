```python
def checkOneAt(s):
    if s.count('@')==1:
        return True
    return False
def checkOneDot(s):
    if s.count('.')==1:
        return True
    return False
def indexOfAt(s):
    return s.rfind('@')
def indexOfDot(s):
    return s.rfind('.')
def checkDifferenceBetweenDotAndAt(s):
    if indexOfDot(s) - indexOfAt(s) > 4:
        return True
    return False
def hasThreePerfixCharacters(s):
    if indexOfAt(s) >= 3:
        return True
    return False
def isSuffixDotCom(s):
    if s[-4:] == ".com":
        return True
    return False
def validateEmail(s):
    if checkOneAt(s) and checkOneDot(s) and checkDifferenceBetweenDotAndAt(s) and hasThreePerfixCharacters(s) and isSuffixDotCom(s):
        return True
    else:
        return False
```

```python
import re

def lenghtOfPassword(password):
    if len(password)>5 and len(password)<16:
        return True
    return False
def upperCase(password):
    if re.search('[A-Z]',password):
        return True
    return False
def lowerCase(password):
    if re.search('[a-z]',password):
        return True
    return False
def specialCharacter(password):
    if re.search('[~!@#$%^&*]',password):
        return True
    return False
def digit(password):
    if re.search('[0-9]',password):
        return True
    return False
def validatePassword(password):
    if lenghtOfPassword(password) and upperCase(password) and lowerCase(password) and specialCharacter(password) and digit(password):
        return True
    return False
```

```python
import pickle
db = {}

def login(email, password):
    db = pickle.load(open("save.txt", "rb"))
    if email in db.keys() and db[email] == password:
        print("Login Successful")
        return
    elif email in db.keys() and db[email] != password:
        print("Incorrect Password")
        return;
    else:
        print("Go for Registration")
        return
    
def register(email, password):
    if validatePassword(password) and validateEmail(email):
        db = pickle.load(open("save.txt", "rb"))
        if email in db.keys():
            print("Go for Login")
            return
        else:
            db[email] = password
            pickle.dump(db, open("save.txt", "wb"))
    else:
        print("Password or Email doesn't meet the given requirements")
while 1:
    print("1. Login")
    print("2. Register")
    print("3. Quit")
    option = int(input("Choice: "))
    if option == 1:
        email = input("Email ID: ")
        password = input("Password: ")
        login(email, password)
    elif option == 2:
        email = input("Email ID: ")
        password = input("Password: ")
        register(email, password)
    else:
        break
    print()
```
